```python
''' 1. Metode yang digunakan Elbow Method
    2. - Menghapus kolom id karena tidak relevan dengan konteksnya
       - Menghitung jumlah kesalahan kuadrat dalam cluster Within Cluster Sum of Squared Errors (WSS) untuk nilai k yang berbeda
       - Memvisualisasikan skor belanja pelanggan dengan pendapatan tahunan  
       - Titik data dipisahkan menjadi 5 kelas yang direpresentasikan dalam warna berbeda

    3. 
    4. Cocok, karena data dapat dikelompokkan terpisah namun memiliki karakteristik serupa(pengelompokkan pelanggan), dan berkemungkinan dapat digunakan 
       untuk kepentingan mall terkait, misalnya menentukan harga produk yang tepat
    5. Atribut yang memiliki korelasi tinggi adalah Customer dan Spending score karena memiliki nilai korelasi 1
    6. Feature yang digunakan adalah Elbow Method
    7. Algoritma yang dipakai adalah KMeans
    8. Hyperparamater berupa df_mall dan wcss
    9. Dataset tersebut merupakan data pembagian pasar ke dalam kelompok pelanggan terpisah yang memiliki karakteristik serupa
    10. Data homogenety dari clustering tersebut adalah Gender
    11. Data heterogenity dari clustering tersebut adalah Spending score
'''

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df_mall = pd.read_csv("/content/drive/My Drive/datasheet/Mall_Customers.csv")
df_mall.drop(["CustomerID"], axis = 1, inplace=True)

from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(df_mall.iloc[:,1:])
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("Within Cluster Sum Of Squares (WCSS)")
plt.show()


plt.figure(figsize=(7,7))
plt.title('Matriks Korelasi Customer Mall')

sns.heatmap(df_mall.corr(), annot=True)
```

```python
km = KMeans(n_clusters=5)
clusters = km.fit_predict(df_mall.iloc[:,1:])
df_mall["label"] = clusters

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_mall.Age[df_mall.label == 0], df_mall["Annual Income (k$)"][df_mall.label == 0], df_mall["Spending Score (1-100)"][df_mall.label == 0], c='blue', s=60)
ax.scatter(df_mall.Age[df_mall.label == 1], df_mall["Annual Income (k$)"][df_mall.label == 1], df_mall["Spending Score (1-100)"][df_mall.label == 1], c='red', s=60)
ax.scatter(df_mall.Age[df_mall.label == 2], df_mall["Annual Income (k$)"][df_mall.label == 2], df_mall["Spending Score (1-100)"][df_mall.label == 2], c='green', s=60)
ax.scatter(df_mall.Age[df_mall.label == 3], df_mall["Annual Income (k$)"][df_mall.label == 3], df_mall["Spending Score (1-100)"][df_mall.label == 3], c='orange', s=60)
ax.scatter(df_mall.Age[df_mall.label == 4], df_mall["Annual Income (k$)"][df_mall.label == 4], df_mall["Spending Score (1-100)"][df_mall.label == 4], c='purple', s=60)
ax.view_init(30, 185)
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```

```python
# Mengimpor library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
 
# Mengimpor dataset
X = df_mall.drop(['Gender'], axis=1)
y = df_mall['Gender']
 
# Membagi data ke dalam training dan test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 99)
 
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() 
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
 
#KNN
from sklearn.neighbors import KNeighborsClassifier
# mengaktikan fungsi klasifikasi untuk knn
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit (X_train,y_train)

y_pred = knn.predict(X_test)
knn.predict_proba(X_test)

from sklearn.metrics import classification_report, confusion_matrix
classification_report(y_test, y_pred)
```
